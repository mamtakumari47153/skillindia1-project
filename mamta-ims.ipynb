```python
import json

data = open("mamta-product.json", 'r')
r = data.read()
data.close()
products = json.loads(r)

```

```python
products
```

```python
prod_id = input("Enter product id: ")
is_billing = False
total_amount = 0
prod_sale_dict = dict()
if not products.get(prod_id):
    print("Invalid product Id!!")
else:
    quant = input("Enter Quantity: ")
    quant = int(quant)
    print(quant)
    if not products[prod_id]["Quantity"] >= quant:
        print('out of stock for this count')
    else:
        name = input("Enter Your name")
        contact = input("Enter your phone number: ")
        products[prod_id]['Quantity'] -= quant
        is_billing = True
        amount = quant * int(products[prod_id]['Price'])
        total_amount += amount
        prod_sale_dict = {
            "Product id": prod_id,
            "Product name": products[prod_id]["Type"],
            "Quantity": quant,
            "Amount": amount
        }
print(prod_sale_dict)
```

```python
if is_billing:
    data = open("Sales.json", 'a')
    js = json.dumps(prod_sale_dict)
    data.write(js)
    data.close()
    data.close()
    print("**************************************************************************************")
    print("Name: ", name)
    print("Phone Number: ", contact)
    print("---------------Billing Product----------------")
    print("Product Name: ", prod_sale_dict['Product name'])
    print("Quantity:", quant)
    print("Amount:", amount)
    print("Total Amount:", total_amount)
    print("---------------Thanks For Coming:)------------")
```
